Problem 3.6 (rainstorm)
(a) Mean $E[X]$:
$$E[X]=\int_0^2 x\frac{x}{8}\,dx+\int_2^8 x\frac{2}{x^2}\,dx
=\frac{1}{8}\left[\frac{x^3}{3}\right]_0^2+2[\ln x]_2^8
=\frac{1}{3}+2\ln4\approx3.106$$

(b) Conditional probability given it started 2 hr ago that it ends within another 1 hr:
$$P(X\le3\mid X>2)=\frac{P(2<X\le3)}{P(X>2)}
=\frac{\int_2^3 \frac{2}{x^2}\,dx}{\int_2^8\frac{2}{x^2}\,dx}
=\frac{\left[-\frac{2}{x}\right]_2^3}{\left[-\frac{2}{x}\right]_2^8}
=\frac{1-\tfrac{2}{3}}{1-\tfrac{1}{4}}=\frac{1/3}{3/4}=\frac{4}{9}\approx0.4444$$

Problem 3.22 (two rivers, A:5 yr, B:10 yr)  
Let $p_A=1/5=0.2,\;p_B=1/10=0.1$ and independence.

(a) Flood in a given year:
$$P(\text{flood})=1-(1-p_A)(1-p_B)=1-0.8\cdot0.9=0.28$$

(b) At least 2 floods in next 5 yrs. $X\sim\mathrm{Binomial}(n=5,p=0.28)$:
$$P(X\ge2)=1-P(X=0)-P(X=1)=1-(0.72)^5-5(0.28)(0.72)^4\approx0.43$$

(c) Reduce annual flood prob $\le0.15$. Test cheapest upgrades that meet $1-(1-p_A')(1-p_B')\le0.15$.
Chosen optimal: upgrade A to 10 yr ($p_A'=0.1$, cost low) and B to 20 yr ($p_B'=0.05$):
$$1-(0.9)(0.95)=0.145\le0.15.$$  
Answer: A to 10 yr and B to 20 yr.

Problem 3.33 (flight delays, exponential mean 0.5 hr so $\lambda=2$)
Tail: $P(X>t)=e^{-2t}$.

(a) John: 3 passengers have 1-hr connection, 7 have 2-hr. Randomly John:
$$P(\text{miss})=\tfrac{3}{10}e^{-2(1)}+\tfrac{7}{10}e^{-2(2)}
=0.3( e^{-2})+0.7(e^{-4})\approx0.053$$

(b) John and Mike (different thresholds 1 hr and 2 hr) but same delay; both miss iff delay $>2$:

(c) Mary waited 0.5 hr already; Johns connection =1 hr. Memoryless property:
$$P(X>1\mid X>0.5)=P(X>0.5)=e^{-1}\approx0.368$$

Problem 3.45 (lognormal recurrence time)

(a) Parameters. Given $E[T]=80$ yr and $c_T=0.40$. For a lognormal variable:
$$\sigma_{\ln T}=\sqrt{\ln(1+c_T^2)}=\sqrt{\ln(1+0.4^2)}\approx0.3853$$
$$\mu_{\ln T}=\ln(E[T])-\tfrac{1}{2}\sigma_{\ln T}^2=\ln(80)-\tfrac{1}{2}(0.3853)^2\approx4.3078$$
(So $ \mu_{\ln T}\approx4.308,\ \sigma_{\ln T}\approx0.385$.)

(b) $P(\text{severe quake within 20 yr})=P(T\le20)$. Use lognormal CDF:
$$P(T\le20)=\Phi\!\Big(\frac{\ln 20-\mu_{\ln T}}{\sigma_{\ln T}}\Big)
=\Phi\!\Big(\frac{\ln20-4.3078}{0.3853}\Big)\approx0.00033.$$

(c) Last quake was 100 yr ago. Want $P(\text{quake in next 1 yr}\mid T>100)$:
$$P(100<T\le101\mid T>100)=\frac{F(101)-F(100)}{1-F(100)}$$
Compute $F(x)=\Phi\!\big((\ln x-\mu_{\ln T})/\sigma_{\ln T}\big)$. Numerically,
$$F(100)\approx0.7799,\; F(101)\approx0.7875$$
so
$$\frac{F(101)-F(100)}{1-F(100)}\approx\frac{0.7875-0.7799}{1-0.7799}\approx0.0344\ (\text{ }3.44\%).$$

Note: It is important that you know the relationship of the mean, the standard deviation, and the coefficeint of variation, between Lognormal and Normal Dsitributions! 

Problem 3.51 (gamma capacity  **answer key formulation**)

Given: capacity $C$ is gamma distributed with mean $2500$ tons and c.o.v. $=0.35$.
For a gamma$(k,\nu)$ with **rate** $\nu$:
$$\frac{k}{\nu}=2500,\qquad \frac{1}{\sqrt{k}}=0.35$$
Hence
$$k=8.16,\qquad \nu=0.00327.$$

(a) Conditional probability:
$$P(C>3000\mid C>1500)=\frac{P(C>3000)}{P(C>1500)}.$$
Using the regularized incomplete gamma function $I_u(\cdot,\cdot)$:
$$P(C>3000)=1-I_u(\nu\cdot3000,k)=1-I_u(9.81,8.16)=0.745,$$
$$P(C>1500)=1-I_u(\nu\cdot1500,k)=1-I_u(4.905,8.16)=0.889.$$
Therefore
$$P(C>3000\mid C>1500)=\frac{0.745}{0.889}=0.838.$$

(b) Probability of no damage over 50 years.
Damage occurs if $C<2000$:
$$P(C<2000)=I_u(\nu\cdot2000,k)=I_u(6.54,8.16)=0.312.$$
Earthquakes of this size follow a Poisson process with rate $1/20$ per year, so
$$\text{mean damaging rate}=0.312\times\frac{1}{20}=0.0156.$$
Thus
$$P(\text{no damage in 50 yr})=\exp(-0.0156\times50)=0.458.$$

(c) Probability at least 4 buildings are damaged in a complex of 4.
Single-building damage probability $=0.312$, survival probability $=0.688$.
$$P(\text{at least 4 damaged})=1-P(\text{none damaged})
=1-(0.688)^4=0.776.$$
